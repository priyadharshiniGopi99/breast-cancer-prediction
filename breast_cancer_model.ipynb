```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
```

```python
data=pd.read_csv(r"C:\Users\priya\Desktop\ipcs\pandas\ML\Excel\breast-cancer.csv")
data.head()
```

```python
data.drop("id",axis=1,inplace=True)
data.head()
```

```python
data.isna().sum()
```

```python
data.info()
```

```python
ind= data.iloc[:,:-1]
dep=data['diagnosis'].map({'M':1,'B':0})
print(ind.shape,dep.shape)
```

```python
ind_train,ind_test,dep_train,dep_test=train_test_split(ind,dep,test_size=0.20,random_state=38)
print(ind_train.shape,dep_train.shape,ind_test.shape,dep_test.shape)
```

```python
sc=StandardScaler()
ind_train=sc.fit_transform(ind_train)
ind_test=sc.transform(ind_test)
```

```python
lr = LogisticRegression()
dt = DecisionTreeClassifier()
rf = RandomForestClassifier()
svc = SVC()
knn = KNeighborsClassifier()
```

```python
lr.fit(ind_train, dep_train)
dt.fit(ind_train, dep_train)
rf.fit(ind_train, dep_train)
svc.fit(ind_train, dep_train)
knn.fit(ind_train, dep_train)
```

```python
models={
    "LogisticRegression":lr,
        "Decision Tree":dt,
        "Random Forest":rf,
        "SVC":svc,
        "KNN":knn
        }
results = {}
trained_models = {}

for name, model in models.items():
    model.fit(ind_train, dep_train)
    trained_models[name] = model
    
    pred = model.predict(ind_test)
    acc = accuracy_score(dep_test, pred)
    results[name] = acc
    
    print(f"{name} Accuracy:", acc)
    
best_algo = max(results, key=results.get)
best_model = trained_models[best_algo]

print("\nBest Algorithm:", best_algo)
print("Best Accuracy:", results[best_algo])
```
